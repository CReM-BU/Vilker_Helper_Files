## Plot gene expression

```python
# You can do this option with plot_two by simply passing in the gene of interest (case sensitive) to the gene parameter as well as the processedData to geneExpressions
fig, ax = plt.subplots(2, 3, figsize=(35,20))
marker_genes = ['Krt8', 'Sfn', 'Lgals3', 'Cldn4', 'Cdkn1a', 'Sprr1a']
axs = ax.flatten()
for i, gene in enumerate(marker_genes):
    SimilarityHelper.plot_two(riemondy.projections["MC-KO"], riemondy.annotations,
             'MC-KO AT1', 'MC-KO AT2',
             gene=gene, geneExpressions=riemondy.processedData, ax=axs[i])
# plt.savefig('/Your/Path/Here.png')
plt.show()
```

## Plot boxplot of desired features of source against relevant features of basis

```python
# First step: Create map of basis labels to the projections of cells with each source label onto said basis labels
riemondySimilarityMap = SimilarityHelper.getMatchingProjections(riemondy)

# Second step: Set basic parameters like plot size and title and generate boxplot
fig, ax = plt.subplots(figsize=(20, 8))
SimilarityHelper.similarityBoxplot(ax, riemondySimilarityMap, title="Riemondy vs MC-KO Similarity Boxplot")
# plt.savefig('/Your/Path/Here.png')
plt.show()
```

```python
# That was quite cluttered! It can help a lot to specify which features you are interested in for the source and basis
riemondyKept = riemondy.toKeep
basisKept = ['MC-KO AT1', 'MC-KO AT2', 'MC-KO Basal', "MC-KO Ciliated", "MC-KO Club"]
# Let's be careful to specify which projections we mean this time, as an scTOP object can have more than one set
riemondySimilarityMap = SimilarityHelper.getMatchingProjections(riemondy, projections=riemondy.projections["MC-KO"], basisKeep=basisKept, sourceKeep=riemondyKept)

# Second step: Set basic parameters like plot size and title and generate boxplot
fig, ax = plt.subplots(figsize=(20, 8))
plt.title("Riemondy vs MC-KO Similarity Boxplot")
SimilarityHelper.similarityBoxplot(fig, ax, riemondySimilarityMap, title="Riemondy vs MC-KO Similarity Boxplot")
# plt.savefig('/Your/Path/Here.png')
plt.show()
```

## Plot over multiple time points

```python
# To do this, you just need to have a time column appropriately identified in your dataset. We'll load such a dataset
bibekObj = TopObject.TopObject("Bibek")
bibekObj.projectOntoBasis(mouseBasis, "MC-KO")
```

```python
# So long as the time information was properly uploaded to the TopObject, you're set to run with no changes
includeCriteria = bibekObj.annotations.isin(bibekObj.toKeep)
fig, axes = SimilarityHelper.plot_two_multiple(
    bibekObj, bibekObj.projections["MC-KO"],
    'MC-KO AT1', 'MC-KO AT2',
    includeCriteria=includeCriteria,
    legendFontSize=12, titleFontSize=36, title="Bibek vs MC-KO Basis Similarity Plots Over Time",
    similarityBounds=(-0.3, 0.8)
)

# plt.savefig('/Your/Path/Here.png')
plt.show()
```
